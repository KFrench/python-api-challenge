```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint 

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "..\python-api-challenge\Resources\weatherpy\cities1.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to the cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
### Perform API Calls
#Perform a weather check on each city using a series of successive API calls.
#Include a print log of each city as it's being processed (with the city number and city name).
```

```python
temp=[]
lat=[]
humi=[]
cloud=[]
wind=[]
long=[]
country=[]
counter = 0
for city in cities: 
    base_url= "http://api.openweathermap.org/data/2.5/weather?"
    units="imperial"
    city_name=city
    #query_url= base_url + "appid=" + weather_api_key + units + "&q=" + city_name
    query_url = f"{base_url}appid={weather_api_key}&units={units}&q={city_name}"
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()
    try:
        temp.append(weather_json['main']['temp'])
        lat.append(weather_json['coord']['lat'])
        long.append(weather_json['coord']['lon'])
        humi.append(weather_json['main']['humidity'])
        cloud.append(weather_json['clouds']['all'])
        wind.append(weather_json['wind']['speed'])
        name= weather_json['name'] 
        country.append(weather_json['sys']['country'])
        counter = counter + 1
        print(f"The city name is:{name} and the number is: {counter}") 
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    
   
   
```

```python
cities_pd=pd.DataFrame(cities)
```

```python
temp_pd=pd.DataFrame(temp)
lat_pd=pd.DataFrame(lat)
long_pd=pd.DataFrame(long)
humi_pd=pd.DataFrame(humi)
cloud_pd=pd.DataFrame(cloud)
wind_pd=pd.DataFrame(wind)
country_pd=pd.DataFrame(country)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
city_all=pd.concat([cities_pd,country_pd,temp_pd,lat_pd,long_pd,humi_pd,cloud_pd,wind_pd],axis=1)
ncolumns=['City', 'Country', 'Temperature(F)', 'Latitude', 'Longitude','Humidity(%)','Cloudiness(%)','Windspeed(mph)']
city_all.columns=ncolumns
```

```python
city_all.dropna(axis=0, how="any", inplace=True)
city_all.to_csv("../Resources/citiesall.csv", index=False, header=True)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
north_pd.dropna()
```

```python
x_axis=north_pd['Temperature(F)']
y_axis=north_pd['Latitude']
```

```python
(slope, intercept,rvalue,pvalue,stderr)=linregress(x_axis,y_axis)
regress_values=x_axis * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_axis,y_axis)
plt.plot(x_axis,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.title("Northern Hemisphere: Max Temp vs. Latitude Linear Regression")
plt.ylabel('Temperature(F)')
plt.xlabel('Latitude')
plt.savefig('NorthTL.png')

plt.show()
```
